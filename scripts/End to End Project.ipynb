```python
df['order_date'] = pd.to_datetime(df['order_date']).dt.normalize()
```

```python
df.head()
```

```python
df['discount_amount'] = df['list_price'] * df['discount_percent'] / 100
```

```python
df['selling_price'] = df['list_price'] - df['discount_amount']
```

```python
df['total_revenue'] = df['selling_price'] * df['quantity']
```

```python
df['total_profit'] = df['total_revenue'] - df['cost_price']*df['quantity']
```

```python
df.head()
```

```python
df['profit_loss_flag']=df['total_profit'].apply(lambda x: 'Profit' if x>0 else('Loss' if x<0 else 'break even'))
```

```python
df['profit_loss_flag'].value_counts()
```

```python
df['total_revenue'].max()
```

```python
df['order_value_bucket'] = pd.cut(
    df['total_revenue'],
    bins=[-1, 5000, 20000, float('inf')],
    labels=['low', 'medium', 'high']
)
```

```python
df['order_value_bucket'].value_counts()
```

```python
df['order_day'] = df['order_date'].dt.day_name()
df['order_month_number'] = df['order_date'].dt.month
df['order_month'] = df['order_date'].dt.month_name()
df['order_year'] = df['order_date'].dt.year
```
